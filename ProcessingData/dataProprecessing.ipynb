### Crop

```python
def save_img_crop(img_path,img_path_save):
    os.chdir(img_path_save)
    print('start save')
    name_iamge = os.listdir(img_path)
    for path_img in name_iamge :
        try : 
          # print(path_img)
          path_x = rf"{path_img}"
          path_images = os.path.join(img_path,path_img)
          img = cv2.imread(path_images)
          img_resize = cv2.resize(img,(108,320))
          # #cắt ảnh
          img_crop = img_resize[15:300, 0:108, :]
          # label_img = 'crop_rota'+path_x[81:]
          # # print(label_img)
          cv2.imwrite(path_img, img_crop)
          # print(label_img)
        except :
            print('error')
    print('done!')
```

```python
def save_img_resize(img_path,img_path_save):

    os.chdir(img_path_save)
    print('start save')
    name_iamge = os.listdir(img_path)
    for path_img in name_iamge :
        try:
         # print(path_img)
         path_x = rf"{path_img}"
         path_images = os.path.join(img_path,path_img)
         img = cv2.imread(path_images)
         img_resize = cv2.resize(img,(108,285))
         # # print(label_img)
         cv2.imwrite(path_img, img_resize)
         # print(label_img)
        except :
            print('error')
    print('done!')
```
